```python
import face_recognition

# Load the two images
image1 = face_recognition.load_image_file("me.jpg")
image2 = face_recognition.load_image_file("me2.jpg")

# Get face encodings
encodings1 = face_recognition.face_encodings(image1)
encodings2 = face_recognition.face_encodings(image2)

# Ensure both images have a detected face
if len(encodings1) == 0 or len(encodings2) == 0:
    print("Face not detected in one of the images.")
else:
    encoding1 = encodings1[0]
    encoding2 = encodings2[0]

    # Compare the two faces
    result = face_recognition.compare_faces([encoding1], encoding2)
    distance = face_recognition.face_distance([encoding1], encoding2)[0]

    print("Do they match?", result[0])
    print("Face distance:", distance)
    # Lower distance means more similar (0.4–0.6 usually means same person)
```

```python
import cv2
import face_recognition
import time

# Load the registered image and encode it
registered_img = face_recognition.load_image_file("images.jpeg")
registered_encodings = face_recognition.face_encodings(registered_img)

if len(registered_encodings) == 0:
    print("No face found in registered image.")
    exit()

registered_encoding = registered_encodings[0]

# Start webcam
cap = cv2.VideoCapture(0)
print("Looking for face... (Press 'q' to quit)")

while True:
    start_time = time.time()

    ret, frame = cap.read()
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Detect face(s)
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

    for (top, right, bottom, left), encoding in zip(face_locations, face_encodings):
        match = face_recognition.compare_faces([registered_encoding], encoding)[0]
        distance = face_recognition.face_distance([registered_encoding], encoding)[0]

        label = "Matched ✅" if match else "Not Matched ❌"
        color = (0, 255, 0) if match else (0, 0, 255)

        # Draw box and label
        cv2.rectangle(frame, (left, top), (right, bottom), color, 2)
        cv2.putText(frame, f"{label} ({distance:.2f})", (left, top - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

        # Stop after first face match attempt
        print(f"{label} | Time: {(time.time() - start_time)*1000:.2f} ms")

    cv2.imshow("Face Match", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
